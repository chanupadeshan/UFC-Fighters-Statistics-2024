<h2 style="font-weight:bold">   Introduction to UFC (Ultimate Fighting Championship) </h2>

The <b> Ultimate Fighting Championship (UFC) </b> stands as the pinnacle of <b> mixed martial arts (MMA) </b>, revolutionizing combat sports since its inception in 1993. By bringing together fighters from diverse backgrounds including <b> boxing, Brazilian Jiu-Jitsu, wrestling, and Muay Thai </b>,the UFC has captured global attention with its adrenaline-fueled events.

With its explosive fights and a roster boasting athletes from across the globe, the UFC has not only redefined sports entertainment but also left an indelible mark on popular culture. Our analysis of <b>UFC fighters' statistics for 2024 </b>will shed light on the trends and performances shaping this dynamic sport.


<h2 style="font-weight:bold">Features Desctiption:</h2>

- **name:** Identifies the UFC fighter.

- **nickname:** If applicable, it's the moniker by which the fighter is known.

- **wins:** Reflects the total victories achieved by the fighter in their career.

- **losses:** Indicates the total number of defeats suffered by the fighter.

- **draws:** Represents the number of matches resulting in a draw for the fighter.

- **height_cm:** Specifies the fighter's height in centimeters.

- **weight_in_kg:** Quantifies the fighter's weight in kilograms.

- **reach_in_cm:** Measures the fighter's arm reach in centimeters.

- **stance:** Describes the fighter's preferred fighting stance (Orthodox/Southpaw/Switch).

- **date_of_birth:** Marks the birthdate of the fighter.

- **significant_strikes_landed_per_minute:** Average rate of significant strikes landed per minute by the fighter.

- **significant_striking_accuracy:** Percentage of successful significant strikes out of total attempts.

- **significant_strikes_absorbed_per_minute:** Average rate of significant strikes absorbed per minute by the fighter.

- **significant_strike_defence:** Percentage of opponent's significant strikes successfully defended by the fighter.

- **average_takedowns_landed_per_15_minutes:** Average number of takedowns successfully executed per 15 minutes.

- **takedown_accuracy:** Percentage of successful takedown attempts out of total attempts.

- **takedown_defense:** Percentage of opponent's takedown attempts successfully defended by the fighter.

- **average_submissions_attempted_per_15_minutes:** Average number of submission attempts made by the fighter per 15 minutes.

### Create two columns : total matches and weight class

```python
##create tptal match column
df["Total matches"] = df["wins"] + df["draws"] + df["losses"]
```

```python
df.info()
```

```python
df["weight_in_kg"] = df["weight_in_kg"].round(1)
```

```python
## function to identify weight class
def weightclass(weight):
    if weight <= 52.2:
        return "Strawweight"
    elif 52.3 <= weight <= 56.1:
        return "Flyweight"
    elif 56.2 <= weight <= 61.2: 
        return "Bantamweight"
    elif 61.3 <= weight <= 65.8:
        return "Featherweight"
    elif 65.9 <= weight <= 70.3:
        return "Lightweight"
    elif 70.4 <= weight <= 77.1:
        return "Welterweight"
    elif 77.2 <= weight <= 83.9:
        return "Middleweight"
    elif 83. <= weight <= 93:
        return "Light Heavyweight"
    elif 93.1 <= weight <= 120.2:
        return "Heavyweight"
```

```python
##create weight class column
df["Weight Class"] = df["weight_in_kg"].apply(weightclass)
```

## **Top 10 players that highest win rate , loss rate and draw rate**

## Win rate

```python
top_10_winning_percentage = df[["name", "wins","Total matches"]]
top_10_winning_percentage["Win Rate"] = df["wins"]/df["Total matches"]*100
highest_winning_percentage = top_10_winning_percentage.sort_values(by=["Win Rate","Total matches"],ascending=False).head(10)
highest_winning_percentage.head(10)
```

```python
#How mappy fighter has 100% win rate
highest_winning_percentage[highest_winning_percentage["Win Rate"] == 100.0].count()
```

## Loss rate

```python
top_10_losses_percentage = df[["name", "losses","Total matches"]]
top_10_losses_percentage["Loss Rate"] = df["losses"]/df["Total matches"]*100
highest_losses_percentage = top_10_losses_percentage.sort_values(by=["Loss Rate","Total matches"],ascending=False).head(10).round(1)
highest_losses_percentage.head(10)
```

### Draw rate

```python
top_10_draw_percentage = df[["name", "draws","Total matches"]]
top_10_draw_percentage["Draw Rate"] = df["draws"]/df["Total matches"]*100
highest_draw_percentage = top_10_draw_percentage.sort_values(by=["Draw Rate","Total matches"],ascending=False).head(10).round(1)
highest_draw_percentage.head(10)
```
